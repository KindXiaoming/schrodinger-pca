```python
import scipy.special
import numpy as np
import matplotlib.pyplot as plt

xs = np.zeros((4,101))
for n in range(4):
    for i in range(-50,51):
        xs[n,i+50] = np.e**(-i**2/(120*np.sqrt(2)))*scipy.special.eval_hermite(n, 1/np.sqrt(60*np.sqrt(2))*i)
```

```python
plt.plot(np.arange(-50,51),xs[1,:])
```

```python
images = []
for i in range(4):
    for j in range(i+1):
        image = xs[i-j,:][:,np.newaxis]*xs[j,:][np.newaxis,:]
        images.append(image)
images = np.array(images)
```

```python
plt.figure(figsize=(20,20))

for i in range(10):
    plt.subplot(1,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(images[i])
    
np.save('./results/fine_se_analytical.npy', images)
#plt.savefig("./fig/fine_se_analytical.pdf", bbox_inches="tight")    
```
